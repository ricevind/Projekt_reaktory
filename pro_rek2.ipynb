```python
# Definicje parametrów
R = 1.905 # cm
L = 9.144 # cm
C_p = 0.8374 #J/g/deg
ro = 1.6*10**-3 #g/cm**3
u_s = 91440 #cm/h średnia prędkości pozorna w reaktorze
ro_b = 1.443 #g/cm**3 gęstść wypełnienia
d_H = -69714 #J/gA entalpia reakcji na gram składnika A
k_r = 0.9353 # H/h/cm/dg przewodnictwo mieszaniny reakcyjnej
D_r = 4645 # cm**2/h współczynnik dyfuzji
C_a0 =3.3*10**-5 #g/cm**3 stężenie wlotowe
T_0 = 316.2 + 273.15 # temp. na wlocie K
T_s = 291 + 273.15 # temp. scianki K
```

```python
# szybkość reakcji w funkcji stopnia przereagowania i temperatury
from numpy import exp

def Ra(f, T):
    ra = 1223040*(1-f)*exp(-(5389/T)) # ga/h/cm**3
    R = -ra*C_a0*ro_b
    return ra
```

```python
# siatka
M = 11 # liczba węzłów wzdłóż promienia
N = 51# liczba węzłów zdłóż długości raktora

dr = R/M
dz = L/N

```

```python
#program - funkcje
# od 1 do L i promien or 1 do M-1
def f_med(fm_l, fml, fmpl, Tml, m):
    sk = (D_r*dz/u_s)*((1/(m*dr)*((fmpl-fml)/dr)+(fmpl-2*fml+fm_l)/dr**2))
    skk = (Ra(fml, Tml)*ro_b*dz)/u_s
    return fml+sk+skk

def T_med(Tm_l, Tml, Tmpl, fml,m):
    sk = (k_r/(ro*C_p*u_s)*(dz/dr**2))*((1/m)*(Tmpl - Tml) + Tmpl - 2*Tml + Tm_l)
    skk = (Ra(fml, Tml)*ro_b*C_a0*d_H*dz)/(ro*C_p*u_s)
    return Tml + sk - skk

# funkcje dla osi

def f_os(f0l, f1l, T0l):
    sk = ((D_r*dz)/(u_s*dr**2))*4*(f1l-f0l)
    skk = (Ra(f0l, T0l)*ro_b*dz)/u_s
    return f0l + sk + skk

def T_os(T0l, T1l, f0l):
    sk = (k_r*4/(ro*C_p*u_s)*dz/dr**2)*(T1l-T0l)
    skk = (Ra(f0l, T0l)*ro_b*C_a0*d_H*dz)/(ro*C_p*u_s)
    return T0l + sk - skk

# funkcje dla scianki

def T_sc(T_s):
    return T_s

def f_sc(fm_2l, fm_1l):
    return 4/3*fm_1l - 1/3*fm_2l
```

```python
# pętle i vectory wyników
# indeksy sa od 0 do M 


# inicializacja macierzy

T = np.zeros((M,N))
f = np.zeros((M,N))

#indeksy maksymalne
m = M - 1 # bo wymiar jest np. 10x10 ale indeksy 0:9 X 0:9
n = N - 1 
# wlot - scianka
T[m,0] = T_s
f[m,0] = 0

# wlot
T[:m,0] = T_0
f[:m,0] = 0

# petla po długości
    #oś
for i in range(0, n): # range jest od 0 do n-1
    f[0, i+1] = f_os(f[0, i], f[1, i], T[0, i])
    T[0, i+1] = T_os(T[0, i], T[1, i], f[0, i])
    # iteracja od osi do scianki
    for j in range(1, m):
        f[j, i+1] = f_med(f[j-1, i], f[j, i], f[j+1, i], T[j, i], j)
        T[j, i+1] = T_med(T[j-1, i], T[j, i], T[j+1, i], f[j, i], j)
    # scianka
    T[m,i+1] = T_sc(T_s)
    f[m,i+1] = f_sc(f[m-2,i], f[m-1,i])
    
```

```python
# obliczenie stężenia Ca

Ca = C_a0*(1-f)
```

Przedstawienie wyników w formie tabelarycznej oraz wykresów

```python
# Tabelka ze stopniem przereagowania
def druk_tablicy(f, lz, sz, plik=sys.stdout):
    print("skrypt {}".format(sys.argv[0]), file=plik)
    print("{:^{sz}}".format("#", sz=sz), end=" ", file=plik)
    for i in range(0,len(f[0:]+1)):
        if i == 10:
            string = "{:^{sz}}".format(i, sz=sz)
            print(string, file=plik)
        else:
            string = "{:^{sz}}".format(i, sz=sz)
            print(string, end=" ", file=plik)

    for j in range(0,len(f.T[0:]+1)):
        print("{:^{sz}}".format(j, sz=sz), end=" ", file=plik)
        for i in range(0,len(f[0:]+1)):
            if i == 10:
                string = "{:*^{sz}.{lz}f}".format(f[i,j], sz=sz, lz=lz)
                print(string, file=plik)
            else:
                string = "{:*^{sz}.{lz}f}".format(f[i,j], sz=sz, lz=lz)
                print(string, end=" ", file=plik)


    
```

```python
# Wykresy przekrojów
def wykresy(tablica):
    dl = len(tablica[0, :])
    dm = len(tablica[:, 0])
    przekroje = [0, dl//4, dl//2, dl//2+dl//4, dl-1]
    x = np.array([i for i in range(0,dm)])
    for i in przekroje:
        plb.plot(x, tablica[:, i], label='przekrój {}'.format(i))
        plt.legend(loc=3)
```
